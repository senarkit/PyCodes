```python
# import necessary packages
import pandas as pd
import numpy as np
import datetime
```

```python
# params - date range
start = '2022-01-24'
days = 35

# Create an array of 5 dates starting at a Monday, one per day
rng = pd.date_range(start, periods=days, freq='D')
df = pd.DataFrame({ 'cal_dt': rng, 'Val' : np.random.randn(len(rng))})
df = df[["cal_dt"]]

# Get week start monday for each date
c = 7
tmp=[]
for i, row in df.iterrows():
    c = c - 1
    if c<0:
        c = 6
    week_dt = row["cal_dt"] + datetime.timedelta(days=(c))
    tmp.append(week_dt)

df["week_dt"] = tmp

# Display
df
```

```python
# params - date range
start = '2022-01-24'
days = 35

# Create an array of 5 dates starting at '2015-02-24', one per day
rng = pd.date_range(start, periods=35, freq='D')
df_dly = pd.DataFrame({ 'cal_dt': rng})
rng = pd.date_range(start, periods=(days/7), freq='W')
df_wk = pd.DataFrame({ 'week_dt': rng})

df = df_dly.merge(df_wk, left_on='cal_dt', right_on='week_dt', how='left')
df
```
